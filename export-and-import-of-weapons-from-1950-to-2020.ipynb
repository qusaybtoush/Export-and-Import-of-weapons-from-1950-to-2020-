```python
#Importing the basic librarires fot analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")  #using style ggplot

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import datetime as dt
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Importing the dataset
df=pd.read_csv("../input/export-and-import-of-weaponsfrom-1950-to-2020/SIPRI ARMS TRANSFER DATA.csv")

# look the data set
df.head()
```

```python
# how much percentage Transfer Type in the dataset

df['Tranfer Type'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',shadow=True)
```

```python
# Creat now columes with name Total

df['Total']=df.sum(axis=1)
df.head()
```

```python
#Make new dataframe for Export & Import

df_ex=df[(df["Tranfer Type"]!="Import")] #filter about Export
df_im=df[(df["Tranfer Type"]=="Import")] #filter about Import
```

```python
# The 10 highest country Export weapons

top_10=df_ex.sort_values(by="Total",ascending=False,ignore_index=True).set_index("Supplier").head(10)
top_10
```

```python
top_10["percentage"]=top_10["Total"]/df['Total'].sum()*100
top_10["percentage"]
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x="Supplier",y="Total",data=top_10)
plt.xticks(rotation=90)
plt.title("The 10 highest country Export weapons ")
plt.show()
```

```python
# The 10 less country Export weapons

less_10=df_ex.sort_values(by="Total",ascending=True,ignore_index=True).set_index("Supplier").head(10)
less_10
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x="Supplier",y="Total",data=less_10)
plt.xticks(rotation=90)
plt.title("The 10 less  country Export weapons ")
plt.show()
```

```python
# The 10 highest country Import weapons

top_im_10=df_im.sort_values(by="Total",ascending=False,ignore_index=True).set_index("Supplier").head(10)
top_im_10
```

```python
top_im_10["percentage"]=top_im_10["Total"]/df['Total'].sum()*100
top_im_10["percentage"]
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x="Supplier",y="Total",data=top_im_10)
plt.xticks(rotation=90)
plt.title("The 10 highest country Import weapons ")
plt.show()
```

```python
# The 10 less country Import weapons

less_im_10=df_im.sort_values(by="Total",ascending=True,ignore_index=True).set_index("Supplier").head(10)
less_im_10
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x="Supplier",y="Total",data=less_im_10)
plt.xticks(rotation=90)
plt.title("The 10 less country Import weapons ")
plt.show()
```
